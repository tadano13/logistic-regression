```python
x = titanic_data.drop(["survived","name","ticket"],axis = 1)
x
```

```python
dummy = pd.get_dummies(x,drop_first=True,dtype = int)
dummy
```

```python
y = titanic_data["survived"].map({0:"Not Survived",1:"Survived"})
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(dummy,y,test_size = 0.2 , random_state = 500 , shuffle = True)
print(f"x_train is {x_train.shape}")
print(f"x_test is {x_test.shape}")
print(f"y_train is {y_train.shape}")
print(f"y_test is {y_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_logr = LogisticRegression()
model_logr
```

```python
model_logr.fit(x_train,y_train)
```

```python
y_pred = model_logr.predict(x_test)
y_pred
```

```python
model_logr.predict([[56, 509900, 3, 1, 0, 0, 0, 0]])
```

```python
model_logr.predict([[36, 5, 1, 1, 1, 0, 0, 0]])
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```
